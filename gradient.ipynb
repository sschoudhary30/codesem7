```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x+3)**2
```

```python
def df(x):
    return 2*(x+3)
```

```python
def gradient(stx, lrate, numit):
    x = stx
    history_x = [x]
    for i in range(numit):
        grad = df(x)
        x = x - lrate * grad
        history_x.append(x)
    return x,history_x
```

```python
x= 2
lrate = 0.1
numit = 50
minx, history_x = gradient(x,lrate,numit)
print(minx)
```

```python
x= np.linspace(-10,10,100)
y=f(x)
plt.plot(x,y)
history_y = [f(i) for i in history_x]
plt.scatter(history_x,history_y)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='f(x) = (x+3)²')

plt.plot(history_x, history_y, 'ro-', label='Gradient Descent Steps')

plt.legend()

plt.grid(True)

print(f"Local minimum found at x = {minx:.4f}")
print(f"f(x) at minimum = {f(minx):.4f}")

```
